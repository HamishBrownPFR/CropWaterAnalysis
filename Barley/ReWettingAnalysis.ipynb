##Bring in libraries and dataframes and set indexes

```python
import datetime
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ETFunctions
%matplotlib inline
```

```python
Data = pd.read_pickle('.\DataFrames\RewettingVWC_df')
```

```python
#Calculate treatment means omitting data prior to 2014-11-05 08:00:00 to avoid NaN values
DataMeans =  Data.ix['2015-02-06 08:00:00':].groupby(level=['Depth','Irrigation', 'Rewet'],axis=1).mean()
DataMeans =  DataMeans.dropna(axis=1) #For some reason it keeps non valid combinations in so need to extract with this function

#Calculate the water content of the soil profile by multiplying the volumetric water content by each layers
#depth and summing.  The 0-15 layers are divided by 2 to average the two readings
ProfileWater = DataMeans.ix[:,'D1I'] * 150/2 + \
               DataMeans.ix[:,'D1B'] * 150/2 + \
               DataMeans.ix[:,'D2'] * 150 + \
               DataMeans.ix[:,'D3'] * 300 + \
               DataMeans.ix[:,'D4'] * 300 + \
               DataMeans.ix[:,'D5'] * 300 + \
               DataMeans.ix[:,'D6'] * 300 + \
               DataMeans.ix[:,'D7'] * 300
                
ProfileWater.ix['2015-02-01':].plot(figsize=(18,18),style=['b-','b--','r-','r--','g-','g--'])
plt.ylabel('Soil Water Content (mm)', fontsize=30);
plt.tick_params(labelsize=25)
plt.title('RainShelter Barley 2014/15', fontsize=28);
```

```python
DailyMax = ProfileWater.resample('D', how='max',axis=0)
DailyDeltas = DailyMax.diff()
DailyDeltas.ix['2015-02-01':].plot(figsize=(18,18),style=['b-','b--','r-','r--','g-','g--'])
plt.ylabel('Soil Water Change (mm)', fontsize=30);
plt.tick_params(labelsize=25)
plt.title('RainShelter Barley 2014/15', fontsize=28);
```

```python
DataMeans2 = Data.ix['2014-11-05 08:00:00':].groupby(level=['Depth', 'Irrigation', 'Rewet','Plot'],axis=1).mean()
DataMeans2 =  DataMeans2.dropna(axis=1)

#Calculate the water content of the soil profile by multiplying the volumetric water content by each layers
#depth and summing.  The 0-15 layers are divided by 2 to average the two readings
ProfileWater2 = DataMeans2.ix[:,'D1I'] * 150/2 + \
               DataMeans2.ix[:,'D1B'] * 150/2 + \
               DataMeans2.ix[:,'D2'] * 150 + \
               DataMeans2.ix[:,'D3'] * 300 + \
               DataMeans2.ix[:,'D4'] * 300 + \
               DataMeans2.ix[:,'D5'] * 300 + \
               DataMeans2.ix[:,'D6'] * 300 + \
               DataMeans2.ix[:,'D7'] * 300

ProfileWater2.ix['2015-02-08':].plot(figsize=(18,18),style=['b-','b-','b-','b-','b--','b--','b--','b--','r-','r-','r-','r-','r--','r--','r--','r--','g-','g-','g-','g-','g--','g--','g--','g--'])
plt.ylabel('Soil Water content (mm)', fontsize=30);
plt.tick_params(labelsize=25)
plt.title('RainShelter Barley 2014/15', fontsize=28);
```

##Graph treatment means by layer

```python
Fig = plt.figure(figsize=(18, 60))

def MakePlot(Position, Depth, LayerThickness, Ymin, Ymax):
        Fig.add_subplot(8,1,Position, color_cycle=['b','b','r','r','g','g'])
        plt.title(Depth + '  VolumetricWaterContent', fontsize=28);
        plt.plot(DataMeans.index,DataMeans.ix[:,Depth]);
        plt.ylabel('mm/mm', fontsize=22);
        plt.tick_params(labelsize=16);
        plt.ylim(Ymin,Ymax);
        return;

MakePlot(1,'D1I',150,0,0.50)
MakePlot(2,'D1B',150,0,0.50)
MakePlot(3,'D2',150,0,0.45)
MakePlot(4,'D3',300,0.05,0.45)
MakePlot(5,'D4',300,0.1,0.45)
MakePlot(6,'D5',300,0.15,0.45)
MakePlot(7,'D6',300,0.2,0.45)
MakePlot(8,'D7',300,0.25,0.45)
```
