```python
OatsLoggedData = pd.read_pickle('.\Oats\DataFrames\TempAndMet.pkl').loc[:,['AboveCanopyPAR','AirTemperature','BelowCanopyPAR']]
OatsAboveCanopy = OatsLoggedData.loc[:,'AboveCanopyPAR'].mean(axis=1).resample('d').sum()
OatsLoggedfPAR = pd.concat([1 - OatsLoggedData.loc[:,'BelowCanopyPAR'].resample('d').sum().divide(OatsAboveCanopy,axis=0)],axis=1,keys=['Oats'],names=['Crop'])
OatsGPs = GH.SetGraphicsPatterns(OatsLoggedData.BelowCanopyPAR.columns,
                    ['-','--'],
                    [0.15,0.45],
                    ['o','^'],
                    ['orange','green','purple'])
```

```python
def MakeTimeSeriesGraphs(Data,name,ymin,ymax):
    Graph = plt.figure(figsize=(18,30))
    Crops = Data.columns.get_level_values(0).drop_duplicates().values
    pos = 1
    for C in Crops:
        GP = globals()[C+'GPs']
        ax = Graph.add_subplot(3,2,pos)
        GH.AddIndexedPlotWithSEM(Data.loc[:,C].dropna(how='all'),ax,GP,2)
        plt.ylabel(name, fontsize=36);
        plt.ylim(ymin,ymax)
        plt.tick_params(labelsize=20)
        plt.title(C, fontsize=28)
        ax.xaxis.set_major_locator(mdates.MonthLocator(interval=int(np.ceil(Data.loc[:,C].dropna(how='all').index.size/30/5))))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))
        GH.AddLegend(4,16,'')
        pos+=1
def MakeSingleCropTimeSeries(Data,Crop):
    graph = plt.figure(figsize=(18,18))
    ax = graph.add_subplot(1,1,1)
    GP = globals()[Crop+'GPs']
    GH.AddIndexedPlot(Data.loc[:,Crop].dropna(how='all'),ax,GP,2)
```

```python
BarleyNDVI = pd.concat([pd.read_excel('.\Barley\RawDatafiles\Canopy.xlsx',sheet_name='NDVI_Raw',index_col=[0],header=[0,1,2])],
                       axis=1,keys=['Barley'],names=['Crop'])
BarleyNDVI.columns = BarleyNDVI.columns.swaplevel('Cultivar','Irrigation')
BarleyNDVI.drop(labels='Mid',level=1,axis=1,inplace=True)
FodderBeetNDVI = pd.concat([pd.read_excel('.\FodderBeet\RawDatafiles\Canopy.xlsx',sheet_name='NDVI_Raw',index_col=[0],header=[0,1,2])],
                       axis=1,keys=['FodderBeet'],names=['Crop'])
#FodderBeetNDVI.drop(labels = pd.to_datetime('2016-11-29'),inplace=True)
OatsNDVI = pd.concat([pd.read_excel('.\Oats\RawDatafiles\Canopy.xlsx',sheet_name='NDVI_Raw',index_col=[0],header=[0,1,2])],
                       axis=1,keys=['Oats'],names=['Crop'])
WheatNDVI = pd.concat([pd.read_excel('.\Wheat\RawDatafiles\Canopy.xlsx',sheet_name='NDVI_Raw',index_col=[0],header=[0,1,2])],
                       axis=1,keys=['Wheat'],names=['Crop'])
NDVI = pd.concat([BarleyNDVI.drop('BS',level=2,axis=1),
                  FodderBeetNDVI.drop('BS',level=2,axis=1),
                  OatsNDVI.drop('BS',level=2,axis=1),
                  WheatNDVI.drop('BS',level=2,axis=1)])
BSNDVI = pd.concat([BarleyNDVI.loc[:,BarleyNDVI.columns.isin(['BS'],level=2)],
                    FodderBeetNDVI.loc[:,FodderBeetNDVI.columns.isin(['BS'],level=2)],
                    OatsNDVI.loc[:,OatsNDVI.columns.isin(['BS'],level=2)],
                    WheatNDVI.loc[:,WheatNDVI.columns.isin(['BS'],level=2)]])
```

## Check out the range in max NDVI and rescale to account for non-zero values with bare soil and sub one values with full cover

```python
for c in Crops:
    NDVImax = np.percentile(NDVI.loc[:,c].dropna(),90,axis=0,).max()
    print(NDVImax)
```

```python
MaxNDVI = pd.DataFrame(index = Crops,columns = ['maxNDVI'],
                       data = [0.85,0.85,0.92,0.85])
#Assume 95th percentile is suitable for barley, fodderbeet and oats as they closed their canopy (see fPARData)
#Assum MaxNDVI needs to be higher than the 95th percentile for wheat because it did not close its canopy
```

```python
for col in BSNDVI.columns:
    CropNDVI = NDVI.loc[:,(col[0],col[1])].min(axis=1)
    BSNDVI.loc[:,col] = [min(CropNDVI[x],BSNDVI.loc[x,col]) for x in BSNDVI.index]
```

```python
Slope = BSNDVI.copy()
Intercept = BSNDVI.copy()
for col in Slope:
    c = col[0]
    Slope.loc[:,col] = [1/(MaxNDVI.loc[c].values[0] -BSNDVI.loc[x,col]) for x in Slope.index]
    Intercept.loc[:,col] = [BSNDVI.loc[x,col] * -Slope.loc[x,col] for x in Intercept.index]
```

```python
ScaledNDVI = NDVI.copy()
for col in ScaledNDVI:
    SlopeCol = (col[0],col[1],'BS','Expt')
    ScaledNDVI.loc[:,col] = [NDVI.loc[x,col] * Slope.loc[x,SlopeCol] + Intercept.loc[x,SlopeCol] for x in ScaledNDVI.index]
    
```

```python
MakeTimeSeriesGraphs(ScaledNDVI,'NDVI',0,1)
```

```python
Graph = plt.figure(figsize=(18,20))
pos = 1
for c in Crops:
    GP = globals()[c+'GPs']
    ax = Graph.add_subplot(2,2,pos)
    gNDVI = ScaledNDVI.loc[Dates.loc[c,'Start']:Dates.loc[c,'Turn'],c]#.groupby(level=[0,1],axis=1).mean()
    gfPAR = LoggedfPAR.loc[:,c].reindex(gNDVI.index)#.groupby(level=[0,1],axis=1).mean()
    #GH.AddIndexedPlot(gfPAR,ax,GP,3)
    GH.AddIndexedXY(gfPAR,gNDVI,ax,GP,15)
    GH.SetAxProps(title = c,titsize=20,ylab='NDVI',xlab='fPAR',
                           labsize=18,ylim=(-.1,1.1),xlim=(-.1,1.1))
    GH.AddLegend(2,18,'')
    plt.plot([0,1],[0,1],'k',lw=3)
    pos += 1
```

```python
Graph = plt.figure(figsize=(18,20))
pos = 1
for c in Crops:
    GP = globals()[c+'GPs']
    ax = Graph.add_subplot(2,2,pos)
    gNDVI = ScaledNDVI.loc[Dates.loc[c,'Start']:Dates.loc[c,'Turn'],c]#.groupby(level=[0,1],axis=1).mean()
    gfPAR = LoggedfPAR.loc[:,c].reindex(gNDVI.index)#.groupby(level=[0,1],axis=1).mean()
    resid = gNDVI - gfPAR
    GH.AddIndexedPlotWithSEM(resid,ax,GP,3)
    GH.SetAxProps(title = c,titsize=20,ylab='NDVI-fPAR',xlab='fPAR',
                           labsize=18,ylim=(-0.5,0.5))
    GH.AddLegend(1,18,'')
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%y'))
    index = resid.dropna(how='all').index
    plt.plot([index.min(),index.max()],[0,0],'k',lw=4)
    pos += 1
```
