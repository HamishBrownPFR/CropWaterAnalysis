##Now lets merge the ryegrass and lucerne data into the same dataframe and transpose so it can be joined with the metadataindex

```python
CoverDataMerged = pd.concat([RyeGrassCoverData, LucerneCoverData], #Date frames to join together
                      join='outer',  #Create a union of the two frames
                      axis=1)       #join then along the column axis
```

```python
CoverDataMerged.axes
```

```python
CoverDataTransposed=CoverDataMerged.transpose()
```

```python
CoverDataTransposed.columns.name='date'
```

```python
CoverDataTransposed.index
```

## Bring in metadata to use for setting multiindex for against plot numbers

```python
MetaData=pd.read_csv('.\IndexFiles\LandPTrialDesign.csv',
                    dtype = {0:np.str} )
```

```python
MetaData
```

#### Set plot column as index

```python
MetaData.index = MetaData['Plot']
```

```python
MetaData.index.name='Plot'
```

```python
MetaData.index
```

## Join cover data and meta data together

```python
CoverDataIndexed = pd.concat([MetaData, CoverDataTransposed], axis=1, sort=True)
```

## Set up index for CoverData

```python
CoverDataIndexed.index.name="Plot"
```

```python
CoverDataIndexed=CoverDataIndexed.set_index(['Plot', 'Species', 'Irrigation', 'Treatment', 'Replicate'])
```

```python
CoverDataIndexed.axes
```

```python
CoverDataIndexed
```

## Transpose CoverData to make date the x axis and make a grouped object to allow calculation of treatment means

```python
CoverData = CoverDataIndexed.transpose()
```

```python
CoverData.axes
```

```python
CoverData.index.name='TIMESTAMP'
```

```python
CoverData[:10]
```

```python
CoverDataGrouped = CoverData.groupby(axis = 1, level=['Irrigation','Species']).mean()
```

## Bring in Metdata and set up accumulated thermal time dataframe

```python
MetData = pd.read_pickle('..\BroadFieldsMet.pkl').resample('D').mean()
```

```python
ThermalTimeAccum = pd.DataFrame(MetData.loc[:,'AirTemperature'].loc['2011-03-10':'2014-04-30'].cumsum())
```

## Add Accumulated thermal time column into CoverData Frame

```python
CoverData.loc[:,'AccumTT'] = ThermalTimeAccum.reindex(CoverData.index).AirTemperature
```
